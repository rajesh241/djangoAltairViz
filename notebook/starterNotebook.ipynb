```python
import altair as alt
import pandas as pd
from altair import datum
alt.renderers.enable("notebook")
source=pd.read_csv("nfhs.csv")
source.head()
```

```python
alt.Chart(source).mark_bar().encode(
y='literacy',
column='state',
x='year:O'
)
```

```python
brush=alt.selection_interval()
base=alt.Chart(source).mark_circle().encode(
x='teenPregnancy',
y='literacy',
color='state',
size='fertilityRate'

).add_selection(
brush
)
addon=alt.Chart(source).mark_bar().encode(
x='mean(marriedBefore18)',
y='year:O'
).transform_filter(
brush
)
base & addon
```
